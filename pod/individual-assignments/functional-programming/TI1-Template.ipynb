````python
# filtra os dados por ano usando o atributo de Data (se o ano do atributo for igual ao ano desejado, retorna o obj)
filtra_por_ano = lambda ano, dados: filter(lambda obj: obj['Data'][0:4] == str(ano), dados)

# calcula o total de gols dos dois times onde usa-se um acumulador que guarda o valor da soma dos atributos de placar
# por fim, retorna o acumulador com o somatorio
conta_gols = lambda dados: reduce(lambda acc, obj: acc + int(obj['Mandante Placar']) + int(obj['Visitante Placar']), dados, 0)

# faz a juncao das duas funcoes anteriores mandando os valores filtrados por ano para a f que conta os gols
# eh essa funcao que ira retornar a tupla (ano, gols)
gols_por_ano = lambda dados: lambda ano: (ano, conta_gols(filtra_por_ano(ano, dados)))

# funcao que sera chamada pelo dict e recebera os dados
# usa uma f auxiliar passando os dados em forma de lista, para que seja possivel realizar a iteracao, e a lista de anos
def contar_gols_por_ano(dados):
    # aqui cria uma copia dos dados como lista para manter a referencia
    # se fizesse direto com o DictReader, como nos exercicios anteriores, o cursor que itera teria que ser resetado (descobri da pior maneira)
    dados = list(dados)
    # chama o map que executa a funcao de contagem dos gols para cada ano
    return map(gols_por_ano(dados), list(range(2000, 2021 + 1)))


mostra_dados("contar_gols_por_ano", "brasi2000-2021.csv", contar_gols_por_ano)
````

## Parte-2: Listas Encadeadas

### Funções Auxiliares

````python
def head(L):
    return L[0]

def tail(L):
    return L[1]


def l2ll(L):
    if not L:
        return None
    else:
        return (L[0], l2ll(L[1:]))
    
def ll2l(LL):
    if not LL:
        return []
    else:
        H = head(LL)
        T = tail(LL)
        return [H] + ll2l(T)
````

**4. Implemente uma função pura `mapLL` de alta ordem, de forma que ela receba e aplique uma função `F` sob cada elemento de uma lista encadeada `LL` que é passada como parâmetro (`mapLL(LL,F)`), retornando uma nova lista encadeada (`LL'`). Depois, use `mapLL` para:**

1. Calcular o fatorial de cada elemento de uma lista encadeada. Use a função `factFun` criada em aula.
2. Converter em string cada elemento de uma lista encadeada.
3. Incrementar +1 cada elemento de uma lista encadeada com uma função `lambda`.

**Resultado Esperado:**
```bash
mapLL-factorial-LL: (1, (24, (5040, (2, (6, (120, (720, None)))))))
mapLL-string-LL: ('1', ('4', ('7', ('2', ('3', ('5', ('6', None)))))))
mapLL-incrementa-LL: (2, (5, (8, (3, (4, (6, (7, None)))))))
```

````python
# funcao de alta ordem que recebe uma funcao que eh aplicada aos itens da lista
def mapLL(funcao, lista):
    lista_aux = []
    
    # funcao auxiliar que sera usada para iterar a lista
    def aux(lista): 
        # caso tenha passado por todos os itens retorna a lista resultante
        if (lista == None):
            return lista_aux
        # se n, adiciona na lista auxiliar o resultado da funcao aplicada ao item atual (head da lista)
        lista_aux.append(funcao(head(lista)))
        # chama a funcao auxiliar novamente, dessa vez passando o tail da lista atual
        # assim, ira caminhar por toda a lista ja que o tail desconsidera o head e este vai mudando a cada iteracao
        return aux(tail(lista))
    
    # por fim, como a lista auxiliar eh uma lista simples, transforma ela no formato original recebido (LL) e retorna
    return l2ll(aux(lista))

# funcao fatorial que multiplica todos os valores de 1 a n
factFun = lambda n: n * factFun(n - 1) if n > 1 else 1


print(f"mapLL-factorial-LL: {mapLL(factFun, LL)}")
print(f"mapLL-string-LL: {mapLL(lambda x: str(x), LL)}")
print(f"mapLL-incrementa-LL: {mapLL(lambda x: x + 1, LL)}")
````

**5. Implemente uma função pura chamada `filterLL` de alta ordem, que recebe uma lista encadeada `LL` e uma função `F`, de modo que produza uma nova lista com cada elemento de `LL` que seja verdade para `F`. Depois use ela para:**

1. Filtrar os elementos ímpares de uma lista encadeada. Usar uma função `lambda`.
2. Filtrar os elementos que são do tipo string de uma lista encadeada. Usar uma função `lambda`.
3. Filtrar os elementos de uma lista encadeada que são primos. Crie uma função usando o paradigma funcional (recursiva e pura) chamada `primo`, que retorne verdadeiro se o número passado como argumento for Primo. Assim, você pode usá-la na chamada da função `filterLL`.

**Resultado Esperado:**
```bash
filterLL-Ímpares-LL (1, (7, (3, (5, None))))
filterLL-Strings-LLX ('7', ('3', ('6', None)))
filterLL-Primos-LL (7, (2, (3, (5, None))))
```

````python
# funcao de alta ordem que recebe uma funcao que eh usada para filtrar os itens da lista
def filterLL(lista, funcao):
    lista_aux = []
    
    # funcao auxiliar que sera usada para iterar a lista
    def aux(lista): 
        # caso tenha passado por todos os itens retorna a lista resultante
        if (lista == None):
            return lista_aux
        
        # se o resultado da funcao for True
        if(funcao(head(lista))):
            # adiciona o valor atual na lista auxiliar
            lista_aux.append(head(lista))
        
        # segue a iteracao chamando a funcao auxiliar novamente, agora com o tail da lista atual (mesma logica de antes)
        return aux(tail(lista))
    
    # transforma a lista auxiliar em LL e retorna
    return l2ll(aux(lista))

# funcao para testar se n eh um numero primo
def primo(n):
    def aux(fator):
        # se n for 1 ou n for maior que 2 e o fator atual eh divisor de n, n nao eh primo
        if(n == 1 or (n > 2 and n % fator == 0)):
            return False
        else:
            # se o fator for menor que a metade de n (ou seja, ainda existem possiveis divisores)
            if(fator < n/2):
                # chama a funcao auxiliar novamente, incrementando o valor que sera testado
                return aux(fator + 1)
            else:
                # basicamente, se n for igual 2 ou terminou a iteracao sem ter algum divisor
                return True
    
    # comeca em 2 porq eh sabido que os divisores de um numero primo sao 1 e ele mesmo
    # logo, ignoro o 1 e soh verifico se ha algum outro entre 2 e n/2
    return aux(2)   

print(f"filterLL-Ímpares-LL: {filterLL(LL, lambda x: x % 2 != 0)}")
print(f"filterLL-Strings-LLX: {filterLL(LLX, lambda x: type(x) is str)}")
print(f"filterLL-Primos-LL: {filterLL(LL, primo)}")
````
