## Programação Funcional vs Manipulação de Dados com Pandas

```python
# modo imperativo
def fat1(n):
    f = 1
    while n > 1:
        f *= n
        n -= 1
    return f
# modo funcional
def fat2(n):
    if n < 2:
        return 1
    else:
        return n * fat2(n-1)

# modo re-uso de funções
from operator import mul
from functools import reduce
def fat3(n):
    return reduce(mul,range(1,n+1))

print("fat1(6)", fat1(6))
print("fat2(6)", fat2(6))
print("fat3(6)", fat3(6))
```

### Aplicação de funções e mapeamento

```python
df_rand = pd.DataFrame(np.random.randn(4,3), columns=list('abc'), 
                       index=['RS', 'SP', 'RJ', 'PR'])

display(df_rand)

np.abs(df_rand)
```

```python
df_rand_abs=df_rand.apply(np.abs)
f = lambda x: x.max() - x.min()
df_rand_abs.apply(f)
```

```python
df_rand_abs.apply(f, axis='columns')
```

```python
def f2(x):
    return pd.Series([x.min(), x.max()], index=['min','max'])

df_rand_abs.apply(f2)
```

```python
form = lambda e: '%.2f' % e

df_rand_abs.applymap(form)
```

```python
df_rand_abs['b'].map(form)
```
