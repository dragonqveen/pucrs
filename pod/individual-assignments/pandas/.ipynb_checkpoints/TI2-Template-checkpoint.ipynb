## 2. Crie um array 2D com a sequência de números ímpares no intervalo 0 a 100 e faça as seguintes operações:
- Média das linhas
- Média das colunas
- Soma dos elementos da última linha
- Soma dos elementos da última coluna

**Resultado Esperado:**
```bash
Média das linhas: [ 5. 15. 25. 35. 45. 55. 65. 75. 85. 95.]
Média das colunas: [46. 48. 50. 52. 54.]

Soma dos elementos da última linha: 475
Soma dos elementos da última coluna: 540
```

````python
impares = np.arange(1, 100, 2).reshape(10, 5)

print(f'Média das linhas: {np.mean(impares, 1)}')
print(f'Média das colunas: {np.mean(impares, 0)}\n')

print(f'Soma dos elementos da última linha: {impares[-1:].sum()}')
print(f'Soma dos elementos da última coluna: {impares[:, -1:].sum()}')
````

**Os 5 primeiros registros do DataFrame:**


````python
info = pd.read_csv('dados/alugueis_info.csv', sep=',', index_col= 0)
valores = pd.read_csv('dados/alugueis_valores.csv', sep=',', index_col= 0)

cols = ['Condominio', 'Aluguel', 'IPTU', 'Seguro Incendio', 'Total']



alugueis = pd.merge(info, valores)

alugueis['Cidade'] = alugueis['Cidade'].astype(bool)
alugueis['Andar'] = alugueis['Andar'].apply(lambda x: 0 if x == '-' else int(x))

def convert(x): 
    return 0. if x in ['Sem info', 'Incluso'] else float(x.replace('R$', '').replace(',', '.'))

alugueis[cols] = alugueis[cols].applymap(convert)

print(alugueis.info())
print(alugueis)
````
