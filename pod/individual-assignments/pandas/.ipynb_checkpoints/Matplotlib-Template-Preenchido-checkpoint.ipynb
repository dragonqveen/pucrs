## Gráfico de Linha ([`plot`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html#matplotlib.axes.Axes.plot))

```python
x_dados = np.arange(40)
y_dados = np.random.randint(low=1, high=20, size=40)
z_dados = np.random.randint(low=4, high=30, size=40)
w_dados = np.random.randint(low=6, high=40, size=40)
```

```python
fig, ax = plt.subplots()

ax.plot(x_dados, y_dados, label='y_dados')
ax.plot(x_dados, z_dados, color='b', label='z_dados')

ax.plot(x_dados, w_dados, color='#444444', linestyle='--', marker='o', label='w_dados')

ax.set_ylabel('RandInt')
ax.set_xlabel('Dados')
ax.set_title('Dados Aleatórios')

ax.legend()

ax.grid(True)


x_tic_data = np.arange(0,40, step=2)
y_tic_data = np.arange(0,40, step=3)

ax.set_xticks(x_tic_data)
ax.set_yticks(y_tic_data)

# plt.style.use('seaborn-deep')

# plt.savefig('plot_nome.pdf')

plt.tight_layout()

print(plt.style.available)
```

## Gráfico de Barras ([`bar`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar.html#matplotlib.axes.Axes.bar), [`barh`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.barh.html#matplotlib.axes.Axes.barh))

```python
fig, ax = plt.subplots()

ax.bar(x_dados, y_dados)

for i in range(len(x_dados)):
    ax.annotate(y_dados[i],
                xy=(x_dados[i], y_dados[i]),
                xycoords='data',
                xytext=(x_dados[i], y_dados[i]),
                textcoords='data',
                horizontalalignment='center',
                verticalalignment='bottom'
               )
plt.show()
```

## Gráfico de Disperção ([`scatter`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.scatter.html#matplotlib.axes.Axes.scatter))

```python
fig, ax = plt.subplots()
ax.scatter(x=x_dados, y=y_dados, 
          s=y_dados*4,
           marker='^'
          )
plt.show()
```

## Gráfico de Pizza ([`pie`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.pie.html#matplotlib.axes.Axes.pie))

```python
fig, ax = plt.subplots()

ax.pie(y_dados[:10], labels=x_dados[:10])

print(y_dados[:10])
print(x_dados[:10])

plt.show()
```

```python
fig, ax = plt.subplots()


posi=np.zeros(10)
posi[2] = 0.1

cores=['blue', 'red', 'lightgrey', 'yellow', 'green', 'gold', 'chocolate', 'lightblue',
       'grey', 'purple']

ax.pie(y_dados[:10], labels=x_dados[:10], explode=posi,
      autopct='%1.2f%%', shadow=True, colors=cores)


plt.show()
```

## Gráfico de Histograma ([`hist`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.hist.html#matplotlib.axes.Axes.hist))

```python
fig, ax = plt.subplots()

ax.hist(y_dados,
        linewidth=0.5,
        edgecolor='white',
#       bins=4,
#       bins=[2,4,8,30]
        orientation='horizontal'
       )

plt.show()
```

## Gráficos de Boxes ([`boxplot`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.boxplot.html#matplotlib.axes.Axes.boxplot))

```python
fig, ax = plt.subplots()

dados = [x_dados, y_dados, z_dados]
dados_l = ['x_dados', 'y_dados', 'z_dados']

ax.boxplot(dados, labels=dados_l,
          #patch_artist=True,
          vert=True,
           showmeans=True,
           showfliers=True,
           notch=True
          )

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Random test data
np.random.seed(19680801)
all_data = [np.random.normal(0, std, size=100) for std in range(1, 4)]
labels = ['x1', 'x2', 'x3']

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

# rectangular box plot
bplot1 = ax1.boxplot(all_data,
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax1.set_title('Rectangular box plot')

# notch shape box plot
bplot2 = ax2.boxplot(all_data,
                     notch=True,  # notch shape
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax2.set_title('Notched box plot')

# fill with colors
colors = ['pink', 'lightblue', 'lightgreen']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

# adding horizontal grid lines
for ax in [ax1, ax2]:
    ax.yaxis.grid(True)
    ax.set_xlabel('Three separate samples')
    ax.set_ylabel('Observed values')

plt.show()
```

## Múltiplos Gráficos em uma Figura ([`subplots`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html#matplotlib.pyplot.subplots))

```python
fig, ax = plt.subplots(2,2)


ax[0,0].set_title('Boxplot')
ax[0,0].boxplot(dados, labels=dados_l,
          #patch_artist=True,
          vert=True,
           showmeans=True,
           showfliers=True,
           notch=True
          )
ax[0,1].set_title('Pie')
ax[0,1].pie(y_dados[:10], labels=x_dados[:10], explode=posi,
      autopct='%1.2f%%', shadow=True, colors=cores)

ax[1,0].set_title('Scatter')
ax[1,0].scatter(x=x_dados, y=y_dados, 
          s=y_dados*4,
           marker='^'
          )

ax[1,1].set_title('Bar')
ax[1,1].bar(x_dados, y_dados)
for i in range(len(x_dados)):
    ax[1,1].annotate(y_dados[i],
                xy=(x_dados[i], y_dados[i]),
                xycoords='data',
                xytext=(x_dados[i], y_dados[i]),
                textcoords='data',
                horizontalalignment='center',
                verticalalignment='bottom'
               )

plt.tight_layout(pad=1.0)

plt.show()
```
