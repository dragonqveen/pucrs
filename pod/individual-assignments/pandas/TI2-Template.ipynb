## 4. Forneça um único DataFrame a partir dos dados contidos em `alugueis_info.csv` e `alugueis_valores.csv`. Depois de corrigir os dados, salve o DataFrame corrigido em um arquivo chamado `alugueis.csv` para reusar nas próximas questões.

_Obs: Para quem usar laços nesta questão, terá um desconto de 25% na nota._


**Resultado Esperado do DataFrame:**
```bash

 #   Column           Non-Null Count  Dtype

```
**Os 5 primeiros registros do DataFrame:**


````python
info = pd.read_csv('dados/alugueis_info.csv', sep=',', index_col= 0)
valores = pd.read_csv('dados/alugueis_valores.csv', sep=',', index_col= 0)

alugueis = info

# transforma em bool
alugueis['Cidade'] = alugueis['Cidade'].astype(bool)
# transforma em int; caso seja um hifen retorna 0
alugueis['Andar'] = alugueis['Andar'].apply(lambda x: 0 if x == '-' else int(x))



def convert(x): 
    return 0. if x in ['Sem info', 'Incluso'] else float(x.replace('R$', '').replace(',', ''))

cols = ['Condominio', 'Aluguel', 'IPTU', 'Seguro Incendio', 'Total']

# converte as colunas de valor monetario para float
alugueis[cols] = valores[cols].applymap(convert)

# remove linhas onde ao menos uma das colunas possui valor null
alugueis = alugueis.dropna()

alugueis.to_csv('dados/alugueis.csv')

print(alugueis.info())
````

## 5.  Forneça um painel de visualização `Axes(2,2)` para os dados de `alugueis.csv` com os seguintes gráficos de pizza em ordem:

- Ocorrência do andar dos imóveis para aluguel.
- Ocorrência da quantidade de quartos dos imóveis para aluguel.
- Ocorrência da quantidade de banheiros dos imóveis para aluguel.
- Ocorrência da quantidade de garagens dos imóveis para aluguel.

**Resultado Esperado**


````python
alugueis = pd.read_csv('dados/alugueis.csv', sep=',', index_col= 0)

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=2, figsize=(15, 15))

def generate_data(column):
    
    agrupado = alugueis.groupby(column)
    count_ordered = agrupado[column].count().sort_values(ascending=False)   
    keys = count_ordered.keys()
    
    return count_ordered, keys

andar_data, andar_keys = generate_data('Andar')

explode = np.zeros(len(andar_keys))
explode[0] = 0.1
    
ax1[0].pie(andar_data, labels=andar_keys, autopct='%1.2f%%', explode=explode)
ax1[0].set_title('Andar dos Imóveis para Aluguel')

quartos_data, quartos_keys = generate_data('Quartos')
ax1[1].pie(quartos_data, labels=quartos_keys, autopct='%1.2f%%')
ax1[1].set_title('Quantidade de Quartos dos Imóveis para Aluguel')

banheiros_data, banheiros_keys = generate_data('Banheiros')
ax2[0].pie(banheiros_data, labels=banheiros_keys, autopct='%1.2f%%')
ax2[0].set_title('Quantidade de Banheiros dos Imóveis para Aluguel')

garagem_data, garagem_keys = generate_data('Vagas Garagem')
ax2[1].pie(garagem_data, labels=garagem_keys, autopct='%1.2f%%')
ax2[1].set_title('Quantidade de Garagens dos Imóveis para Aluguel')

plt.show() 
````

## 6. Para a base de dados `alugueis.csv`, apresente um gráfico de barras com a média do valor dos aluguéis relativo ao número de quartos para os imóveis com e sem mobilia. 

**Resultado Esperado:**


````python
alugueis = pd.read_csv('dados/alugueis.csv', sep=',', index_col= 0)

#qtde_quartos = sorted(alugueis['Quartos'].unique())
qtde_quartos = np.arange(1, 11)

# separa os dados em 2 grupos, um com mobilia outro sem
agrupados_mobiliado = alugueis.groupby(alugueis['Mobiliado'] == True)

# atribui o grupo sem/com mobilia na respectiva variavel
com_mobilia = agrupados_mobiliado.get_group(True)
sem_mobilia = agrupados_mobiliado.get_group(False)

alugueis_com = com_mobilia.groupby('Quartos')['Aluguel'].mean()
alugueis_sem = sem_mobilia.groupby('Quartos')['Aluguel'].mean()

a = alugueis_com.to_frame(name="Com").reset_index()
b = alugueis_sem.to_frame(name="Sem").reset_index()

a = a.merge(pd.DataFrame(qtde_quartos, columns=['Quartos']), how='outer')
dataframe_alugueis = a.merge(b, how='outer')

fig, ax = plt.subplots(figsize=(15, 7))
width = 0.25

step = np.arange(len(qtde_quartos))

barra_com_mobilia = ax.bar(step - width/2, dataframe_alugueis['Com'], width, label='Com Mobília')
barra_sem_mobilia = ax.bar(step + width/2, dataframe_alugueis['Sem'], width, label='Sem Mobília')

ax.set_ylabel('Valor em Reais')
ax.set_title('Média do Valor dos Aluguéis')
ax.set_xlabel('Número de Quartos')
ax.set_xticks(step, qtde_quartos)

ax.legend(loc="upper right")

ax.bar_label(barra_com_mobilia, padding=3, fmt='%.2f')
ax.bar_label(barra_sem_mobilia, padding=3, fmt='%.2f')

plt.show()
````

## 7. Para a base de dados `alugueis.csv`, apresente dois gráficos `boxplot`  (`Axes(1,2)`) do valor de `IPTU`, sendo um para os 3 maiores e outro para os 3 menores alugueis em média relativo a quantidade de banheiros.

_Dica: nesta questão calcula-se a média do aluguel por quantidade benheiros para saber quem são os 3 maiores e menores aluguéis. Depois, plota-se o IPTU para cada um deles._

**Resultado Esperado:**



````python
alugueis = pd.read_csv('dados/alugueis.csv', sep=',', index_col= 0)

# agrupa por qtde de banheiros e para a coluna aluguel aplica a media; depois transforma a coluna de IPTU em lista
agrupados = alugueis.groupby(['Banheiros']).agg({'Aluguel':'mean', 'IPTU': list})

agrupados = agrupados.sort_values('Aluguel', ascending=False)

maiores = agrupados.head(3)
menores = agrupados.tail(3).sort_values('Aluguel')

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

fig.suptitle('IPTU dos 3 Maiores e Menores Aluguéis Médio por Quantidade de Banheiros')

ax[0].boxplot(maiores['IPTU'].values, showmeans=True, showfliers=False)
ax[0].set_title('IPTU dos 3 Maiores')
ax[0].set_ylabel('Valor em Reais')
ax[0].set_xlabel('Quantidade de Banheiros')
ax[0].set_xticklabels(labels=maiores['IPTU'].keys(), rotation=45)

ax[1].boxplot(menores['IPTU'].values, labels=menores['IPTU'].keys(), showmeans=True, showfliers=False)
ax[1].set_title('IPTU dos 3 Menores')
ax[1].set_ylabel('Valor em Reais')
ax[1].set_xlabel('Quantidade de Banheiros')
ax[1].set_xticklabels(labels=menores['IPTU'].keys(), rotation=45)

plt.show()
````
